```python
# Option 1: Use Ollama on Colab (lightweight)
# Ollama will run on Colab's free GPU

import subprocess
import time

# Install Ollama
!curl -fsSL https://ollama.ai/install.sh | sh

# Start Ollama in background
subprocess.Popen(['ollama', 'serve'], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
print("Waiting for Ollama to start...")
time.sleep(5)

# Download model
!ollama pull mistral  # Fast, lightweight model
print("✅ Ollama ready on http://localhost:11434")
```

```python
# Run the blog generator
from backend.pipeline import create_pipeline
from backend.models import GraphState

# Create pipeline
pipeline = create_pipeline(mode="open_book")

# Define your blog topic
topic = "The future of AI in education"

# Run the generator
initial_state = GraphState(
    topic=topic,
    mode="open_book",
    errors=[]
)

result = pipeline.invoke(initial_state)

# Display the generated blog
print("\n" + "="*80)
print("GENERATED BLOG")
print("="*80)
print(result.get("blog", "No blog generated"))
```

```python
# Save the blog to a file
with open("generated_blog.md", "w") as f:
    f.write(result.get("blog", ""))

print("✅ Blog saved to generated_blog.md")

# Download from Colab Files
from google.colab import files
files.download('generated_blog.md')
```
